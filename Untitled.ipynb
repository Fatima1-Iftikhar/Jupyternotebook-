```python
# put text on image

font = cv2.FONT_HERSHEY_SIMPLEX
org = (50, 50)
fontScale = 1
color = (255, 0, 0)
thickness = 2
image1 = cv2.putText(image, 'OpenCV', org, font, 
                   fontScale, color, thickness, cv2.LINE_AA)


plt.subplot(1,1,1)
plt.imshow(image1)    #Display an image
plt.show()
```
